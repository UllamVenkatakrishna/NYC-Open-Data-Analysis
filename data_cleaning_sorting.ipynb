```python
# Fill missing values in numerical columns with the mean
for column in ts_df.select_dtypes(include=np.number):
  ts_df[column].fillna(ts_df[column].mean(), inplace=True)

# Fill missing values in categorical columns with the mode
for column in ts_df.select_dtypes(include='object'):
  ts_df[column].fillna(ts_df[column].mode()[0], inplace=True)

print(ts_df.isnull().sum())
```

```python
# Fill missing values in numerical columns with the mean
for column in pk_df.select_dtypes(include=np.number):
  pk_df[column].fillna(pk_df[column].mean(), inplace=True)

# Fill missing values in categorical columns with the mode
for column in pk_df.select_dtypes(include='object'):
  pk_df[column].fillna(pk_df[column].mode()[0], inplace=True)

print(pk_df.isnull().sum())
```

```python
print("Column names for rt_df:")
print(rt_df.columns.tolist())
print("\nColumn names for ts_df:")
print(ts_df.columns.tolist())
print("\nColumn names for pk_df:")
print(pk_df.columns.tolist())
```

```python
common_columns = set(rt_df.columns) & set(ts_df.columns) & set(pk_df.columns)
print("Common columns in all three DataFrames:")
print(list(common_columns))
```

```python
# Convert 'OPT_Code' to a consistent data type (e.g., string) in all DataFrames before merging
rt_df['OPT_Code'] = rt_df['OPT_Code'].astype(str)
ts_df['OPT_Code'] = ts_df['OPT_Code'].astype(str)
pk_df['OPT_Code'] = pk_df['OPT_Code'].astype(str)

merged_df = rt_df.merge(ts_df, on=['School_Year', 'OPT_Code'])
merged_df = merged_df.merge(pk_df, on=['School_Year', 'OPT_Code'])
```

```python
# Filter for routes with a specific number
filtered_df = merged_df[merged_df['Route_Number'] == 123]

# Filter for schools with a specific name
filtered_df = merged_df[merged_df['School_Name'] == 'Central High']
filtered_df
```

```python
grouped_df = merged_df.groupby('Route_Number')['Number_of_Riders'].sum().reset_index()
grouped_df
```

```python
grouped_df = merged_df.groupby('OPT_Code')[['Door_To_Door_Service', 'Stop_To_School_Service', 'Common_Carrier_Svc_Metrocards']].sum().reset_index()
grouped_df
```
